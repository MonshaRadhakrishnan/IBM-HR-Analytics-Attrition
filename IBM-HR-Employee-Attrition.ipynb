```python
df['Age'].describe().reset_index().transpose()
```

```python
agegrp = []
for i in df['Age']:
    if 0<i<=18:
        agegrp.append('< 18')
    elif 18<i<=25:
        agegrp.append('19 to 25')
    elif 25<i<=35:
        agegrp.append('26 to 35')
    elif 35<i<=45:
        agegrp.append('36 to 45')
    else:
        agegrp.append('46+')
```

```python
df['AgeGrp'] = agegrp
```

```python
df.head()
```

```python
df.where(df['Attrition']==1).groupby('AgeGrp')['Attrition'].value_counts().sort_values()
```

```python
## Age 26 to 35 --> Attrition rate is more
```

```python
df.groupby('BusinessTravel')['Attrition'].value_counts(normalize=True)*100
```

```python
## Attrition rate of People who travel frequently 24% more
```

```python
df.groupby('DailyRate')['Attrition'].value_counts(normalize=False)
```

```python
df['DailyRate'].describe().reset_index().transpose()
```

```python
drgrp = []
for i in df['DailyRate']:
    if 0<i<=102:
        drgrp.append('< 102')
    elif 102<i<=400:
        drgrp.append('103 to 400')
    elif 400<i<=700:
        drgrp.append('401 to 700')
    elif 700<i<=1100:
        drgrp.append('701 to 1100')
    elif 1100<i<=1300:
        drgrp.append('1101 to 1300')
    else:
        drgrp.append('1300+')
```

```python
df['drgrp'] = drgrp
```

```python
df.head()
```

```python
dr = pd.DataFrame(df.groupby('drgrp')['Attrition'].value_counts())
dr['%'] = df.groupby('drgrp')['Attrition'].value_counts(normalize=True)*100
dr
```
